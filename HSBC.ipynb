```python
import requests
import urlparse
from bs4 import BeautifulSoup
```

```python
request_get = requests.get('https://www.hsbc.com.tw/1/2/Misc/common-tw/premier-faq')
response = request_get.text.encode('utf8')
soup = BeautifulSoup(response)
getdata = soup.select('li , p , h3')
len(getdata)
fid = open('premier-faq.txt','w')
for i in range(0,len(getdata)):
    question = getdata[i]
    fid.write(question.prettify("utf8"))
    print question
fid.close()
```
